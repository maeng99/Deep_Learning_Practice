```python
from sklearn.datasets import load_iris, load_diabetes

iris_data=load_iris()
diabetes_data=load_diabetes()
```

```python
print(iris_data.target)
```

```python

plt.scatter(iris_data.data[:,3], iris_data.target)
```

```python
X_data = diabetes_data.data
y_data = diabetes_data.target
plt.scatter(X_data[:,3], y_data)
plt.show()
```

```python
x=X_data[:,2]
```

```python
plt.scatter(x, y_data)
plt.show()
```

```python
w=1
b=1
y_pred=x[0]*w+b
print(y_pred)
```

```python
plt.scatter(x, y_data)
start=(-0.1, -0.1*w+b)
end=(0.2, 0.2*w+b)
plt.plot([start[0],end[0]],[start[1],end[1]],color='red')
plt.show()
```

```python
for i in range(10):
  for x_i,y_i in zip(x,y_data):
    y_pred = x_i * w+b
    err=y_i-y_pred
    w_rate=x_i
    w+=w_rate*err
    b+=1*err
print(w,b)
```

```python
plt.scatter(x, y_data)
start=(-0.1, -0.1*w+b)
end=(0.2, 0.2*w+b)
plt.plot([start[0],end[0]],[start[1],end[1]],color='red')
plt.show()
```

```python
for idx in range(len(X_data[0])):
  plt.scatter(X_data[:,idx],y_data)
  plt.show()
```

```python
error_track=np.array([])

for i in range(100):
  total_error=np.array([])
  for x_i,y_i in zip(x, y_data):
    y_pred=x_i*w+b
    err=y_i-y_pred
    total_error=np.append(total_error, err)
  total_error=abs(total_error).mean()
  error_track=np.append(error_track, total_error)
print(w,b)
```

```python
plt.plot(error_track)
```
